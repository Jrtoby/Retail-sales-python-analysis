## Visualizations

- Sales by Region

- Monthly Sales Trend

```python
# EXPLORATORY DATA ANALYSIS (EDA) / BUSINESS ANALYSIS
# Total Sales
df["Sales"].sum()
```

```python
# Total Profit
df["Profit"].sum()
```

```python
# Top Region by Sales
df.groupby("Region")["Sales"].sum()
```

```python
#Lowest Region by sales
df.groupby("Region")["Sales"].sum().idxmin()
```

```python
# Top category by sales
df.groupby("Category")["Sales"].sum().idxmax()
```

```python
# Chart Sales by Region
import matplotlib.pyplot as plt
df.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# MONTHLY SALES TREND
# Extract the year-month, then add up all sales for each month
df["Month"]=df["Order Date"].dt.to_period("M")
df.groupby("Month")["Sales"].sum()
```

```python
# Chart
df.groupby("Month")["Sales"].sum().plot(kind="line")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# TOP 10 PRODUCTS BY SALES
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Chart
df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10).plot(kind="barh")
plt.title("Top 10 Products by sales")
plt.xlabel("Sales")
plt.show()
```

```python
# Most profitable category
df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
```

```python
#Loss making products
df[df["Profit"]<0][["Product Name","Profit"]].head(10)
```

```python
# Segment Contribution
df.groupby("Segment")["Sales"].sum()
```

## Key Insights

- The West region contributed the highest sales performance.
- Technology is the most profitable category.
- Some products generate losses despite sales.
- The business generated strong overall sales with positive profitability.
- sales Trend Analysis (
1. Overall sales show an upward trajectory from 2014 to 2017, indicating business growth over time.
2. Revenue becomes significantly stronger in 2017 compared to earlier years.
3. Several sharp peaks suggest seasonal demand.)

## Recommendations
